```python
import matplotlib.pyplot as plt
import numpy as np

# Create the data
x = np.array([2010, 2015, 2020, 2025, 2030])
y = np.array([100, 200, 300, 400, 500])

# Plot the data
plt.plot(x, y)

# Set the title and labels
plt.title("STEM-Driven Healthcare Wealth Creation")
plt.xlabel("Year")
plt.ylabel("Wealth Created (Billions of USD)")

# Show the plot
plt.show()
```
